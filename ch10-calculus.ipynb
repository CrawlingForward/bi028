## 1. Limit (极限)

### $\S$Squeeze Theorem (夹逼定理)

#### Example: 

$$\lim_{x \to 0} \frac{3\sin^2 x}{4x^2}$$

When $x \in (0,\epsilon)$, $\sin x < x < \tan x \Rightarrow 1 < \frac{x}{\sin x} < \frac{1}{\cos x} \Rightarrow \cos x < \frac{\sin x}{x} < 1$,

Since $\lim_{x \to 0} \cos x = 1$, thus
$$
\lim_{x \to 0} \frac{\sin x}{x} = 1
$$
Therefore,
$$
\lim_{x \to 0} \frac{3\sin^2 x}{4x^2} = \frac{3}{4}
$$

## 2. logarithm function

```python
import math
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

x = np.arange(0.05, 3, 0.05)
y1 =[math.log(a, 1.5) for a in x]
plt.plot(x, y1, linewidth=2, color="#007500", label="$\log_{1.5} x$")
y2 =[math.log(a, 2) for a in x]
plt.plot(x, y2, linewidth=2, color="#9F35FF", label="$\log_2 x$")
y3 =[math.log(a, 3) for a in x]
plt.plot(x, y3, linewidth=2, color="#F75000", label="$\log_3 x$")
plt.legend(loc=0)
plt.grid(True)
plt.plot([1, 1], [y1[0], y1[-1]], 'r--', linewidth=2)
```

## Logarithm curves

```python
from math import log
import matplotlib.pyplot as plt
%matplotlib inline

x = [i/100.0 for i in range(1, 300)]
y = map(log, x)
fig, ax = plt.subplots(figsize=(5, 4))
ax.plot(x, y, 'r-', linewidth=3, label='log curve')
a = [x[20], x[175]]
b = [y[20], y[175]]
ax.plot(a, b, 'g-', linewidth=2, label='straight line')
ax.plot(a, b, 'b*', markersize=10, alpha=.75, label='points')
ax.legend(loc=0, frameon=True, fancybox=False, framealpha=0, markerfirst=False, markerscale=1)
ax.grid(True)
ax.set_xlabel('$x$')
ax.set_ylabel('$log x$')
```

## The Directional Derivatives (方向导数)

Let $f(x)=f(x_1, x_2, \cdots, x_n)$ be a scalar function, its __directional derivative__ along $v=(v_1, \cdots, v_n)$ can be defined as
$$
\left(\frac{\partial f}{\partial x_1}, \cdots, \frac{\partial f}{\partial x_n}\right) \cdot v
$$

## Gradient (Jacobian, 梯度)

The direction $v$ with the maximum directional derivative is the gradient of $f$, which can be denoted as $\nabla f(x)$.

## Uniform Distribution and Central Limit Theorem (CLT, 中心极限定理)

```python
import numpy as np
u = np.random.uniform(0.0, 1.0, 10000)
fig, ax = plt.subplots(2,2, figsize=(6,8))
ax[0,0].hist(u,bins=80,facecolor='g',alpha=.75)
ax[0,0].grid(True)

times = 10000
for time in range(times):
	u += np.random.uniform(0.0, 1.0, 10000)
u /= times
ax[1,0].hist(u, 80, facecolor='g', alpha=.75)
ax[1,0].grid(True)

n = np.random.normal(0.0, 1.0, 10000)
ax[0,1].hist(n,bins=80,facecolor='g',alpha=.75)
ax[0,1].grid(True)

times = 10000
for time in range(times):
	n += np.random.normal(0.0, 1.0, 10000)
n /= times
ax[1,1].hist(n, 80, facecolor='g', alpha=.75)
ax[1,1].grid(True)
```

## Curve fitting with polynomial fitting

```python
import numpy as np
import pylab as pl
%pylab inline
```

```python
np.random.seed(100)
x = np.random.random(20)
y = np.sin(x**2)

def plot_polyfit(x, y, degree=1):
    xfit = np.linspace(0, 1, 10000)
    p = np.polyfit(x, y, degree)
    yfit = np.polyval(p, xfit)
    pl.scatter(x, y, c='k')
    pl.plot(xfit, yfit)
    pl.xlabel('$x$')
    pl.ylabel('$y$')

plot_polyfit(x, y, 3)
```
